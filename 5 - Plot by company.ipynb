# 5. Plot by company 

## Import data and packages

```python
import pandas as pd
month_groupings = pd.read_pickle('articles_proprocessed_for_visualization.pkl')
```

```python
from ipywidgets import interact
from bokeh.io import push_notebook, show, output_notebook
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
from bokeh.layouts import column, row

output_notebook()
```

## Prepare plots & updates for dynamic graph

Note that each of the topics is represented by an individual graph. Thus, we have to supress the toolbar which would otherwise be displayed for each individual graph (which of course would look very messy).

```python
def plot_topic(topic_number):
    plot = figure(x_axis_type='datetime', y_axis_label='values', y_range=(0,1), plot_width=450, 
                  plot_height=125, toolbar_location=None, title = 'Topic {}'.format(topic_number))
    graph = plot.vbar(x='date', top=topic_number, width = 1, line_width = 3, 
                      source=ColumnDataSource(data=month_groupings.loc[('American Greetings Corp')]))
    return plot, graph
```

```python
plot1, graph1 = plot_topic('1')
plot2, graph2 = plot_topic('2')
plot3, graph3 = plot_topic('3')
plot4, graph4 = plot_topic('4')
plot5, graph5 = plot_topic('5')
plot6, graph6 = plot_topic('6')
plot7, graph7 = plot_topic('7')
plot8, graph8 = plot_topic('8')
plot9, graph9 = plot_topic('9')
plot10, graph10 = plot_topic('10')
```

```python
def update(Company):
    source = ColumnDataSource(data=month_groupings.loc[Company])
    graph1.data_source.data = source.data
    graph2.data_source.data = source.data
    graph3.data_source.data = source.data
    graph4.data_source.data = source.data
    graph5.data_source.data = source.data
    graph6.data_source.data = source.data
    graph7.data_source.data = source.data
    graph8.data_source.data = source.data
    graph9.data_source.data = source.data
    graph10.data_source.data = source.data
    push_notebook()
```

## Initiate graph

We display the dropdown menu above the graph. That is not standard practice, but it works and looks better, all we need to do is supress a warning.

```python
import warnings
warnings.filterwarnings('ignore')

interact(update, Company = month_groupings.index.levels[0].tolist())
column1 = column(plot1, plot3, plot5, plot7, plot9)
column2 = column(plot2, plot4, plot6, plot8, plot10)
show(row(column1, column2), notebook_handle=True);
```
